Importando as bibliotecas

```python
import numpy as np # Biblioteca de estatística
import matplotlib.pyplot as plt # Biblioteca de visualizacao de dados
import pandas as pd # Biblioteca de base de dados, ETL, consumo, etc
```

Importando e consolidando as bases de dados

```python
Trimeste_012023 = pd.read_csv('https://raw.githubusercontent.com/pcmassonjr/ProjetoAplicado/main/Base%20de%20Dados/glp-2023-01.csv', sep=';') 
Trimeste_022023 = pd.read_csv('https://raw.githubusercontent.com/pcmassonjr/ProjetoAplicado/main/Base%20de%20Dados/glp-2023-02.csv', sep=';')
Consolidado_2023 = pd.concat([Trimeste_012023, Trimeste_022023])
df_Consolidado2023 = Consolidado_2023
```

Verificando os tipo de dados da base consolidada

```python
df_Consolidado2023.dtypes
```

Limpando as linhas totalmente vazias

```python
df_Consolidado2023.dropna(how= 'all')
```

Alterando o tipo de dado da coluna "Data da Coleta" de String para o tipo Datetime

```python
df_Consolidado2023['Data da Coleta'] = pd.to_datetime(df_Consolidado2023['Data da Coleta'], format='%d/%m/%Y')
```

Criando uma coluna de AnoMes

```python
df_Consolidado2023['AnoMes'] = df_Consolidado2023['Data da Coleta'].dt.strftime('%Y%m')
```

Verificando tipo de dados após ajustes

```python
df_Consolidado2023.dtypes
```
