Importacao das bibliotecas

```python
import numpy as np # Biblioteca de estatística
import matplotlib as mp # Biblioteca de visualizacao de dados
import pandas as pd # Biblioteca de base de dados, ETL, consumo, etc
```

Importacao da base de dados

```python
Trimeste_012023 = pd.read_csv('https://raw.githubusercontent.com/pcmassonjr/ProjetoAplicado/main/Base%20de%20Dados/glp-2023-01.csv', sep=';') 
Trimeste_022023 = pd.read_csv('https://raw.githubusercontent.com/pcmassonjr/ProjetoAplicado/main/Base%20de%20Dados/glp-2023-02.csv', sep=';')
Consolidado_2023 = pd.concat([Trimeste_012023, Trimeste_022023])
```

Visulaizando a Base de dados

```python
display(Consolidado_2023)
```

Criacao dos cálculos

```python
Valor_Venda_Minimo = np.min(pd.to_numeric(Consolidado_2023['Valor de Venda'], errors='coerce'))
Valor_Venda_Medio = np.mean(pd.to_numeric(Consolidado_2023['Valor de Venda'], errors='coerce'))
Valor_Venda_Maximo = np.max(pd.to_numeric(Consolidado_2023['Valor de Venda'], errors='coerce'))
display(Valor_Venda_Minimo,Valor_Venda_Medio,Valor_Venda_Maximo)
```
